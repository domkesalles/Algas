```python
import re
from textblob import TextBlob
import GetOldTweets3 as got

# FunÃ§Ã£o para limpar o texto do tweet
def clean_tweet_text(text):
    # Remover caracteres especiais e hiperlinks
    cleaned_text = re.sub(r'http\S+|[^a-zA-Z\s]', '', text)
    return cleaned_text.strip()

# ConfiguraÃ§Ãµes da pesquisa
query = "felps"
count = 100  # NÃºmero mÃ¡ximo de tweets para recuperar

# CriaÃ§Ã£o do objeto de pesquisa
tweetCriteria = got.manager.TweetCriteria().setQuerySearch(query).setMaxTweets(count)

# Executa a pesquisa
tweets = got.manager.TweetManager.getTweets(tweetCriteria)

# AnÃ¡lise de sentimento
for tweet in tweets:
    cleaned_text = clean_tweet_text(tweet.text)
    analysis = TextBlob(cleaned_text)
    polarity = analysis.sentiment.polarity
    if polarity > 0:
        sentiment = "positivo"
    elif polarity < 0:
        sentiment = "negativo"
    else:
        sentiment = "neutro"

    print(f"Tweet: {cleaned_text}")
    print(f"Sentimento: {sentiment}")
    print()
```
